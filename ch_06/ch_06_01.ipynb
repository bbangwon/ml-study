# 파이플롯

- 파이플롯
  - 맷플롯립을 이용할 때 가장 기본이 되는 객체
  - 파이플롯(pypolt) 위에 그림(figure) 객체를 올리고 그 위에 그래프에 해당하는 축(axes)을 올림
  - 그림 위에 축을 열러장 올리면 여러 개의 그래프 작성
  - X 객체와 Y 객체 값 쌍으로 좌표평면 위에 점을 찍음
  - plot 함수로 점들을 연결


````python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X, Y)
````

- pyplot 객체 내부에 있는 하나의 그림 객체 위에 코사인 그래프와 사인 그래프를 그림


````python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]
X_2 = range(100)
Y_2 = [np.sin(value) for value in X]
plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
````

# 그림과 축

- 그림과 축
  - 그림은 그래프를 작성하는 밑바탕이 됨
  - 축은 실제로 그래프를 작성하는 공간


````python
fig, ax = plt.subplots()  # (1) figure와 axes 객체 할당

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

ax.plot(X_1, Y_1)  # (2) plot 함수를 사용하여 그래프 생성
ax.set(
    title="cos graph", xlabel="X", ylabel="Y"
)  # 그래프 제목, X축 라벨, Y축 라벨 설정

plt.show()  # 그래프 출력
````

````python
fig = plt.figure()  # figure 반환
fig.set_size_inches(10, 10)  # figure의 크기 지정
ax_1 = fig.add_subplot(1, 2, 1)  # 첫번째 그래프 생성
ax_2 = fig.add_subplot(1, 2, 2)  # 두번째 그래프 생성
ax_1.plot(X_1, Y_1, c="b")  # 첫 번째 그래프 설정
ax_2.plot(X_2, Y_2, c="g")  # 두 번째 그래프 설정
plt.show()
````

# 서브플롯 행렬

- 서브플롯 행렬
  - 축을 여러 개 만들 때 서브플롯으로 축 객체 공간 확보
  - 그림 객체에서 add_subplot 함수 사용
  - 또는 plot 객체에서 subplots 함수 사용


````python
fig, ax = plt.subplots(nrows=1, ncols=2)
print(ax)
````

````python
print(type(ax))
````

- #(2) subplots 함수에서 2x2 행렬 그림 객체가 생성되어 ax 변수에 4개의 축 객체가 2x2 넘파이 배열 형태로 들어가 있음
- 넘파이 배열의 인덱스로 각 축 객체에 접근하여 그래프를 생성


````python
x = np.linspace(-1, 1, 100)  # (1) x값과 y_n값 생성
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)  # (2) 2x2 figure 객체를 생성
ax[0, 0].plot(x, y_1)  # (3) 첫 번째 그래프 생성
ax[0, 1].plot(x, y_2)  # (4) 두 번째 그래프 생성
ax[1, 0].plot(x, y_3)  # (5) 세 번째 그래프 생성
ax[1, 1].plot(x, y_4)  # (6) 네 번째 그래프 생성
plt.show()
````

- 행과 열을 지정하고, 세 번째 숫자는 축의 위치


````python
ax1 = plt.subplot(321)  # (1) 첫 번째 공간에 axes 생성
plt.plot(x, y_1)
ax2 = plt.subplot(322)  # (2) 두 번째 공간에 axes 생성
plt.plot(x, y_1)
ax3 = plt.subplot(312)  # (3) 두 번째 공간에 axes 생성
plt.plot(x, y_1)
ax4 = plt.subplot(325)  # (4) 다섯 번재 공간에 axes 생성
plt.plot(x, y_1)
ax5 = plt.subplot(326)  # (5) 여섯 번째 공간에 axes 생성
plt.plot(x, y_1)
plt.show()
````
