# 시본

- 시본(seaborn)
  - 맷플롯립을 바탕으로 다양한 함수 사용을 돕는 일종의 래퍼(wrapper) 모듈
  - 맷플롯립과 동일한 결과물이 나오며, 작성 과정이 간단
  - 그림 객체나 축 객체 같은 복잡한 개념이 없음
  - xticks 설정하지 않아도 각 축에 라벨 자동으로 생성
  - 데이터프레임과 x, y에 해당하는 열 이름만 지정하면 됨


```python
import numpy as np  # numpy 모듈 호출
import pandas as pd  # pandas 모듈 호출
import matplotlib.pyplot as plt  # matplotlib 모듈 호출
import seaborn as sns  # (1) seaborn 모듈 호출

fmri = sns.load_dataset("fmri")  # (2) fmri 데이터셋 사용
sns.set_style("whitegrid")  # (2) 기본 스타일 적용
sns.lineplot(x="timepoint", y="signal", data=fmri)  # (4) 선 그래프 작성
```

- fmri 데이터는 연속형 값 외에도 다양한 범주형 값 가짐
- 이럴 때 맷플롯립으로 표현하기는 상당히 복잡하고, 시본은 hue 매개변수만 추가하면 그래프 그릴 수 있음


```python
fmri.sample(n=10, random_state=1)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
```

# 회귀 그래프

- 회귀 그래프(regression plot)
  - 회귀식을 적용하여 선형회귀 추세선을 그래프에 함께 작성
  - 선형회귀 추세선
    - 데이터를 기반으로 데이터의 x값 대비 y값 변화를 예측하는 직성
  - 함수 regplot 사용


- 매개변수 x_ci는 신회구간의 비율을 나타냄


```python
tips = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips, x_ci=95)
```

- x,y를 기준으로 데이터의 분포 표현
- 함수 scatterplot 사용


```python
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

# 비교 그래프

- 비교 그래프(counter plot)
  - 범주형 데이터의 항목별 개수


```python
tips = sns.load_dataset("tips")
sns.countplot(x="smoker", hue="time", data=tips)
```

# 막대그래프

- 막대그래프
  - y값이 연속형 값일 경우 해당 값들의 평균을 나타냄
  - 데이터의 신뢰구간을 검은색 막대로 표현
  - 함수 barplot 사용


```python
sns.barplot(x="day", y="total_bill", data=tips)
```
