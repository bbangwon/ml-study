```python
import pandas as pd

data_url = (
    "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
)
df_data = pd.read_csv(data_url, sep="\\s+", header=None)
df_data.columns = [
    "CRIM",
    "ZN",
    "INDUS",
    "CHAS",
    "NOX",
    "RM",
    "AGE",
    "DIS",
    "RAD",
    "TAX",
    "PTRATIO",
    "B",
    "LSTAT",
    "MEDV",
]  # 데이터 열 이름 지정
df_data.head()
```

- 판다스로 데이터 다루기
  - 데이터에 수식 적용하기
    - 데이터가 벡터가 아닌 매트릭스 형태라서 수식을 변형
      - y = w1x1 + w2x2 + ... + w13x13 + w0 \* 1
    - 두 벡터의 곱셈 연산을 벡터의 내적으로 나타낼 수 있음
    - 전체 데이터를 X로, 가중치를 W로 나타냄


```python
import numpy as np

df_data["weight_0"] = 1  # weight 0값 추가
df_data = df_data.drop("MEDV", axis=1)  # Y값 제거
df_matrix = df_data.values  # 행렬(matrix) 형태로 변환
weight_vector = np.random.random_sample((14, 1))  # 가중치 w 생성
df_matrix.dot(weight_vector)  # 내적 연산 실행 결과 출력
```

- 결과값은 임의로 나옴
